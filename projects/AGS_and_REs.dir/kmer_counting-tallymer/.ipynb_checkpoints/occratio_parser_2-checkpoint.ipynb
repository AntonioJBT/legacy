```python
# distribution of non unique mers (counting each non unique mer more than once)

def nonUniqueMersMore(infile):

    path = os.getcwd()
    
    with open(infile, 'r') as f:

        kmerValues = []

        while True:
            line = f.readline()
#            print line
            if not line:
                break
            
            if re.match(r'(.*) non unique mers (.*)(\\*) more than once(.*)(\\*)', line):
#                print line
                kmerValues.append(line)
                
                while True:
                    values = f.readline()
                    if not line:
                        break
                    if values[0] != '#':
#                        print values
                        kmerValues.append(values)
                        
                        with open(str('nonUniqueMersMore_'+str(infile)), 'a+') as newFile:
                            newFile.write(values)

                    else:
                        break

        return kmerValues
```
