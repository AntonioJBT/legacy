```python
import pandas
import sqlite3

input_file = 'JULIAN_sheet_1.csv'
plink_missing_code = '-99999999'
db_name = 'BESTD_csvdb'

df = pandas.read_csv(input_file)

#Missing: join pt_id with sample id for genotypes and GEx, then create plink pheno file. 

#Prepare file for alternate phenotype analysis for plink using named variables and headers:
#http://pngu.mgh.harvard.edu/~purcell/plink/data.shtml#pheno
df2 = df.copy()
df2.insert(0, 'FID', '-99999999')
df2.insert(1, 'IID', '-99999999')

#All phenotypes must be numerically coded, including missing values, in the alternate phenotype file. 
#The default missing value is -9, set to -99999999, use  --missing-phenotype in plink to specify this:
df2.fillna('-99999999')

#Create SQL database and insert tables:
df.to_sql(name='JULIAN_sheet_1_original', if_exists='replace', flavor='sqlite', con=sqlite3.connect('BESTD_csvdb'))
df2.to_sql(name='JULIAN_sheet_1_for_plink', if_exists='replace', flavor='sqlite', con=sqlite3.connect('BESTD_csvdb'))
```
